```python
import mlflow
```

```python
MLFLOW_EXP_NAME = "churn-xgboost"
MODEL_DIR       = "artifacts"
RANDOM_STATE    = 42
```

```python
os.makedirs(MODEL_DIR, exist_ok=True)
mlflow.set_experiment(MLFLOW_EXP_NAME)
```

```python
experiment_configs = [
    {"n_estimators": 100, "max_depth": 3, "learning_rate": 0.1,  "subsample": 0.8},
    {"n_estimators": 200, "max_depth": 5, "learning_rate": 0.05, "subsample": 0.8},
    {"n_estimators": 150, "max_depth": 4, "learning_rate": 0.08, "subsample": 0.9},
]

best_f1      = 0.0
best_model   = None
best_metrics = {}
best_run_id  = None
```

```python
from xgboost import XGBClassifier
```

```python
for i, params in enumerate(experiment_configs, 1):
    print(f"\n{'─'*50}")
    print(f" MLflow Run {i}/3 → xgboost-run-{i}")
    print(f" Params: {params}")

    with mlflow.start_run(run_name=f"xgboost-run-{i}") as run:

        # ── Train ──────────────────────────────────────
        model = XGBClassifier(
            **params,
            eval_metric  = "logloss",
            random_state = RANDOM_STATE,
            verbosity    = 0,
        )
        model.fit(X_train, y_train,
                  eval_set=[(X_test, y_test)],
                  verbose=False)

        # ── Evaluate ───────────────────────────────────
        y_pred  = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:, 1]

        metrics = {
            "accuracy"  : round(accuracy_score(y_test, y_pred),  4),
            "precision" : round(precision_score(y_test, y_pred), 4),
            "recall"    : round(recall_score(y_test, y_pred),    4),
            "f1_score"  : round(f1_score(y_test, y_pred),        4),
            "roc_auc"   : round(roc_auc_score(y_test, y_proba),  4),
        }

        print(classification_report(y_test, y_pred,
                                    target_names=["No Churn", "Churn"]))
        print(f" Metrics: {metrics}")

        # ── Log to MLflow ──────────────────────────────
        mlflow.log_params(params)
        mlflow.log_metrics(metrics)
        mlflow.sklearn.log_model(model, "model")

        # ── Track best ─────────────────────────────────
        if metrics["f1_score"] > best_f1:
            best_f1      = metrics["f1_score"]
            best_model   = model
            best_metrics = metrics
            best_run_id  = run.info.run_id

print(f"\n{'='*50}")
print(f" Best Run ID : {best_run_id}")
print(f" Best F1     : {best_f1:.4f}")
print(f"{'='*50}")
```

```python
with open(f"{MODEL_DIR}/model.pkl",    "wb") as f: pickle.dump(best_model, f)
with open(f"{MODEL_DIR}/encoders.pkl", "wb") as f: pickle.dump(le, f)
with open(f"{MODEL_DIR}/scaler.pkl",   "wb") as f: pickle.dump(scaler, f)
with open(f"{MODEL_DIR}/metrics.json", "w")  as f: json.dump(best_metrics, f, indent=2)

print(" Artifacts saved to → artifacts/")
print("   model.pkl    — best XGBoost model")
print("   encoders.pkl — LabelEncoder for categoricals")
print("   scaler.pkl   — StandardScaler for numerics")
print("   metrics.json — best run metrics")
```
